# Julia 機器學習：GLM 線性迴歸

## 作業 027：波士頓房價預測資料集

請使用 GLM 中的模型，建立一個預測模型來預測波士頓的房價。

```julia
using GLM, RDatasets, MLDataUtils
```

## 讀取資料

```julia
boston = dataset("MASS", "Boston")
```

```julia
show(boston[1:5, :], allcols=true) #查看所有欄位名稱
```

```julia
#將資料切割成訓練資料與測試資料
indecies = MLDataUtils.shuffleobs(collect(1:nrow(boston))) #將原資料的順序洗牌
train_ind, test_ind = MLDataUtils.splitobs(indecies, at = 0.8); #0.8的資料當訓練資料，0.2的資料當測試資料
```

```julia
train = boston[train_ind, :] #建立訓練用資料的DataFrame
test = boston[test_ind, :] #建立測試用資料的DataFrame
```

```julia
#依據訓練用資料建立出模型
model = GLM.lm(@formula(MedV ~ Crim + Zn + Indus + Chas + NOx + Rm + Age + Dis + Rad + Tax + PTRatio + Black + LStat), train)
```

```julia
#將測試用的資料放入模型，得到預測結果
predict(model, test)
```

```julia
#計算模型的R-squared值(越接近1表示資料越接近線性模型)
GLM.r²(model)
```

```julia
#計算模型的調整後R-squared值
GLM.adjr²(model)
```
